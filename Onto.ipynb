```python
# Teste de hiperparametros. Alt taxa de aprendizado.

import tensorflow as tf
from tensorflow.keras import layers, models

def build_model(learning_rate=0.01, num_layers=2, units=64):
    model = models.Sequential()
    model.add(layers.Input(shape=(784,))) # Exemplo para dados achatados (como imagens 28x28)
    
    # Adiciona camadas conforme o parâmetro
    for i in range(num_layers):
        model.add(layers.Dense(units, activation='relu'))
    
    model.add(layers.Dense(10, activation='softmax')) # 10 classes de saída
    
    # Compilação com o Learning Rate que vamos testar
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer,
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalização e achatamento (Flatten)
x_train, x_test = x_train / 255.0, x_test / 255.0
x_train = x_train.reshape((-1, 784))
x_test = x_test.reshape((-1, 784))
```

```python
print("Iniciando Teste B: Learning Rate excessivo...")
model_b = build_model(learning_rate=1.0) # 1.0 é extremamente alto para o Adam

history_b = model_b.fit(x_train, y_train, 
                        epochs=5, 
                        batch_size=128, 
                        validation_split=0.2, 
                        verbose=1)

# Verificação de falha
if history_b.history['accuracy'][-1] < 0.20:
    print("\n[ALERTA] Falha detectada: O modelo não convergiu (Learning Rate muito alto).")
```

```python
#Executando o Experimento de Batch Size Subótimo: batch_size=1
# Cenário 1: Batch Size Subótimo (Muito Pequeno)
BATCH_SIZE = 1
LEARNING_RATE = 0.01 # Um valor que funcionaria bem com um batch maior

print(f"Iniciando Teste: Batch Size = {BATCH_SIZE}")

# Usando a função de build que definimos antes
model_batch = build_model(learning_rate=LEARNING_RATE)

# Vamos rodar poucas épocas, pois batch_size=1 é muito lento
history_batch = model_batch.fit(x_train, y_train, 
                                epochs=50, 
                                batch_size=BATCH_SIZE, 
                                validation_split=0.2, 
                                verbose=1)

# Análise de Instabilidade
loss_values = history_batch.history['loss']
is_unstable = any(abs(loss_values[i] - loss_values[i-1]) > 1.0 for i in range(1, len(loss_values)))

if is_unstable or history_batch.history['accuracy'][-1] < 0.5:
    print("\n[ALERTA MLTO] Falha detectada: Instabilidade de Gradiente / Suboptimal Batch Size.")
```

```python
from owlready2 import *

# 1. Inicialização da Ontologia
onto = get_ontology("http://www.ime.usp.br/~alexandre/mlto#")

with onto:
    # --- TAXONOMIA (Classes) ---
    class Execution(Thing): pass
    class State(Thing): pass
    class Symptom(Thing): pass
    
    class Infected(State): pass
    class Healthy(State): pass
    class GradientInstability(Symptom): pass

    # --- RELAÇÕES (Propriedades) ---
    # Adicionamos 'FunctionalProperty' para permitir o uso do sinal de '='
    class has_state(ObjectProperty, FunctionalProperty): 
        domain = [Execution]
        range = [State]
        
    class has_symptom(ObjectProperty): # Um erro pode ter vários sintomas, então não é funcional
        domain = [Execution]
        range = [Symptom]

    # --- ATRIBUTOS (Dados) ---
    class accuracy(DataProperty, FunctionalProperty): 
        domain = [Execution]
        range = [float]
        
    class batch_size(DataProperty, FunctionalProperty): 
        domain = [Execution]
        range = [int]

    # --- FUNÇÃO DE MAPEAMENTO CORRIGIDA ---
    def mapear_resultado_para_mlto(id_exp, b_size, acc_final):
        exp = Execution(id_exp)
        exp.batch_size = b_size
        exp.accuracy = acc_final
        
        if b_size == 1:
            # Agora o '=' funciona porque definimos como FunctionalProperty
            exp.has_state = Infected(f"Infected_State_{id_exp}")
            # Como has_symptom NÃO é funcional, usamos .append() ou uma lista []
            exp.has_symptom.append(GradientInstability(f"Gradient_Noise_{id_exp}"))
            print(f">>> {id_exp}: Detectada INFECÇÃO por instabilidade de gradiente.")
        else:
            exp.has_state = Healthy(f"Healthy_State_{id_exp}")
            print(f">>> {id_exp}: Processo SAUDÁVEL.")

# 2. Execução
mapear_resultado_para_mlto("EXP_001_STOCHASTIC", 1, 0.48)
mapear_resultado_para_mlto("EXP_002_MINIBATCH", 50, 0.97)

# 3. Salvamento
onto.save(file="tese_mlto_alexandre.owl", format="rdfxml")

print("\n✅ Ontologia salva com sucesso e erro de FunctionalProperty resolvido!")
```

```python
# Consultando a ontologia populada
print("🔍 PESQUISANDO EXPERIMENTOS INFECTADOS NA MLTO:\n")

# Buscando todos os indivíduos da classe Execution que possuem o estado Infected
for exp in onto.Execution.instances():
    # Verificando se o estado dele é uma instância de Infected
    if isinstance(exp.has_state, onto.Infected):
        print(f"⚠️ Alerta de Falha!")
        print(f"   Experimento: {exp.name}")
        print(f"   Acurácia: {exp.accuracy}")
        print(f"   Sintomas Detectados: {[s.name for s in exp.has_symptom]}")
        print("-" * 30)

# Verificando o total de triplos gerados (opcional, para métricas da tese)
print(f"\nTotal de instâncias de Execução: {len(onto.Execution.instances())}")
```
